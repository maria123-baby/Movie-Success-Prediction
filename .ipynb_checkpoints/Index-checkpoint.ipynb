```python
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

movie_data = pd.read_csv("movies/imdb_top_1000.csv")
selected_features=['Series_Title','Certificate','Genre','IMDB_Rating','Overview','Director','Star1','Star2','No_of_Votes','Gross']
#replacing null values with null string
for feature in selected_features:
    movie_data[feature] = movie_data[feature].fillna('')
combined_features = movie_data['Series_Title']+' '+movie_data['Certificate']+' '+movie_data['Genre']+' '+movie_data['Overview']+' '+movie_data['Director']
print(combined_features)

tfidf = TfidfVectorizer(stop_words='english')
feature_vectors = tfidf.fit_transform(combined_features)

#compute similarity matrix
similarity_matrix = cosine_similarity(feature_vectors)



# Example usage

```

```python
def get_similar_movies(genre, overview, plot, director, top_n=5):
    # Create a single input vector
    input_text = genre + ' ' + overview + ' ' + plot + ' ' + director
    input_text_vector = tfidf.transform([input_text]).toarray()
    # Normalize numeric inputs
   # input_numeric = scaler.transform([[release_year, runtime]])
    
    # Combine text and numeric inputs
    input_vector = np.hstack(input_text_vector).reshape(1, -1) 

    # Calculate similarity
    similarity_scores = cosine_similarity(input_text_vector,feature_vectors).flatten()

    # Get top N similar movies
    sorted_indices = similarity_scores.argsort()[::-1]
    top_movies = [movie_data['Series_Title'][i] for i in sorted_indices[:top_n]]

    return top_movies
def search():
    similar_movies = get_similar_movies(
    genre="Action Sci-Fi",
    overview="A mind-bending thriller",
    plot="A skilled thief steals secrets through dreams",
    director="Christopher Nolan",
    top_n=10
    )
    for movies in similar_movies:
        print(movies)
```
